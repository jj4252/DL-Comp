```python
from pathlib import Path
import pandas as pd
```

```python
iNat_dir = Path('data/iNat_birds/birds_train_small')
NA_species_csv = Path('aba_english_scientific.csv')
```

```python
NA_species_df = pd.read_csv(NA_species_csv)
NA_sci_names = set(NA_species_df['scientific_name'].apply(lambda x: x.lower()))
```

```python
def get_scientific_name(dir_name):
   parts = dir_name.split('_')
   name = f"{parts[-2]} {parts[-1]}"
   return name.lower()
```

```python
iNat_species = set(get_scientific_name(d.name) for d in iNat_dir.iterdir() if d.is_dir())

iNat_NA_species = iNat_species.intersection(NA_sci_names)
iNat_non_NA_species = iNat_species.difference(NA_sci_names)
```

```python
print(f"Number of iNat species: {len(iNat_species)}")
print(f"Number of NA species: {len(NA_sci_names)}")
print(f"Number of iNat NA species: {len(iNat_NA_species)}")
print(f"Number of iNat non-NA species: {len(iNat_non_NA_species)}")
```
